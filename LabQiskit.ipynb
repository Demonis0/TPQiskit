```python
 # Import Qiskit
from qiskit import QuantumCircuit
# Create quantum circuit
circ = QuantumCircuit(3,1)
circ.h(0)
circ.z(1)
circ.y(2)
circ.cx(0,2)
circ.swap(1,2)
circ.measure(0,0)
# Draw the circuit
circ.draw(output="mpl")
```

```python
from qiskit import QuantumRegister, ClassicalRegister
# Create registers
q_a = QuantumRegister(4,name="ctl")
q_b = QuantumRegister(3,name="work")
q_c = QuantumRegister(1,name="target")
# Create quantum circuit
c4X = QuantumCircuit(q_a,q_b,q_c)
c4X.ccx(q_a[0],q_a[1],q_b[0])
c4X.ccx(q_a[2],q_b[0],q_b[1])
c4X.ccx(q_a[3],q_b[1],q_b[2])
c4X.cx(q_b[2],q_c[0])
c4X.ccx(q_a[3],q_b[1],q_b[2])
c4X.ccx(q_a[2],q_b[0],q_b[1])
c4X.ccx(q_a[0],q_a[1],q_b[0])
# Draw the circuit
c4X.draw(output="mpl")
```

```python
from qiskit.circuit.library.standard_gates import YGate
# Create quantum circuit
cY = QuantumCircuit(3)
c3Y_gate = YGate().control(2)
cY.append(c3Y_gate,[0,2,1])
cY.draw('mpl')
```

```python
from qiskit.circuit.library.standard_gates import HGate
# Create quantum circuit
ch = QuantumCircuit(4)
c3h_gate = HGate().control(3, ctrl_state="100") #first number is last qubit!
ch.append(c3h_gate,[0,2,3,1]) #first three numbers indicate controls
ch.draw('mpl')
```

```python
import numpy as np
# Import Qiskit
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, plot_state_city
import qiskit.quantum_info as qi
# Create circuit
qsim = QuantumCircuit(2,2)
qsim.h(0)
qsim.h(1)
qsim.measure(0,0)
qsim.measure(1,1)
print(qsim)
# Transpile for simulator
simulator = AerSimulator()
qsim = transpile(qsim, simulator)
# Run and get counts
result = simulator.run(qsim,shots=1000).result()
counts = result.get_counts(qsim)
plot_histogram(counts, title='check Hadamard transform')
```

```python
# Create circuit
qubitorder = QuantumCircuit(5)
qubitorder.x(0)
qubitorder.x(2)
qubitorder.measure_all()
print(qubitorder)
# Transpile for simulator
simulator = AerSimulator()
qubitorder = transpile(qubitorder, simulator)
# Run and get counts
result2 = simulator.run(qubitorder,shots=1000).result()
counts2 = result2.get_counts(qubitorder)
plot_histogram(counts2, title='Order of qubits')
```

```python
# Import Qiskit
from qiskit import QuantumCircuit
from qiskit.circuit.library.standard_gates import SGate, TGate
# Create quantum circuit
circ = QuantumCircuit(3)

circ.h(0)

s1_gate = SGate().control(1)
circ.append(s1_gate,[1,0])

T1_gate = TGate().control(1)
circ.append(T1_gate,[2,0])

circ.h(1)

s2_gate = SGate().control(1)
circ.append(s2_gate,[2,1])

circ.h(2)

circ.swap(0,2)

circ.draw(output="mpl")
```
